**Cuales son los 5 pokemones con el HP mas alto?**

```python
pokemon_df.sort_values(by='hp', ascending=False).head()
```

```python
pokemon_df['hp'].mean()
```

**Cuales son los 5 pokemones con el attack mas alto?**

```python
pokemon_df.sort_values(by='attack', ascending=False).head()
```

```python
pokemon_df['attack'].mean()
```

**Cuales son los 5 pokemones con el defense mas alto?**

```python
pd.set_option('display.max_columns', None)
pokemon_df.sort_values(by='defense', ascending=False).head()
```

```python
pokemon_df['defense'].mean()
```

**Que tipo de pokemones son los mas comunes?**

```python
pokemon_df['type1'].value_counts().head()
```

**Cuantos tipos de pokemon existen?**
