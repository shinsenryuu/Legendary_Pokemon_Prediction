## Preguntas iniciales

**indexar segun listado**

```python
pokemon_df.set_index('id', inplace=True)
```

```python
pokemon_df.head()
```

**Reemplazar valores NaN**

```python
pokemon_df['percent-male'] = pokemon_df['percent-male'].fillna(value=pokemon_df['percent-male'].mean())
pokemon_df['percent-female'] = pokemon_df['percent-female'].fillna(value=pokemon_df['percent-female'].mean())
pokemon_df = pokemon_df.fillna(value='None')
pokemon_df.head()
```

**Eliminar columna total**

```python
pokemon_df.drop('total', axis=1, inplace=True)
pokemon_df
```

## Regresion Lineal

```python
pokemon_df.dtypes
```

**Crear clasificacion de pokemones legendarios**

```python
sns.pairplot(pokemon_df)
```

```python
pokemon_df['legendary'] = pokemon_df.legendary.astype(bool)
pokemon_df['legendary'].dtype
```

```python
sns.heatmap(pokemon_df.corr(), annot=True)
```

```python
x = pokemon_df[[ 'hp', 'attack', 'legendary', 'spattack', 'spdefense', 'speed','percent-male', 'percent-female']]
y = pokemon_df['defense']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,x.columns,columns=['Coeficiente'])
coeff_df
```

## Predicciones

```python
predictions = lm.predict(x_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions),bins=50);
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
acc_test = lm.score(x_test, y_test)
print acc_test 
```
